```python
## HUE SHIFTS

fig, axs = plt.subplots(2, 2, figsize=(15, 10), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = 0.2, wspace=0.15)
axs = axs.ravel()
# fig.suptitle("Layers/AAP 2D: " + str(layers_dict[n_exp][0]) + "/" + str(layers_dict[n_exp][2][0]) +
#              ", Layers/AAP 3D: " + str(layers_dict[n_exp][1]) + "/" + str(layers_dict[n_exp][2][1]), fontsize = 15)
hue_shift = 270
for i, hue_shift in enumerate([0, 90, 180, 270]):
    pretrained = True
    pt = "(Pretrained) " if pretrained else ""
    filename = "pt_5k_hueshift_" + str(hue_shift) if pretrained else "5k_hueshift_" + str(hue_shift)
    x = np.arange(0, 5, 1)
    y2_mean, y2_std, y3_mean, y3_std = np.loadtxt("Experiments/cmnist/plots/best/" + filename + ".txt")

    axs[i].set_xticks(ticks = list(range(5)), labels=[0, 12, 60, 120, 1000000])

    axs[i].plot(x, y2_mean, lw=1, label='2D')
    axs[i].plot(x, y3_mean, lw=1, label='3D')
    axs[i].fill_between(x, np.add(y2_mean, y2_std), np.subtract(y2_mean, y2_std), alpha=0.2)
    axs[i].fill_between(x, np.add(y3_mean, y3_std), np.subtract(y3_mean, y3_std), alpha=0.2)
    axs[i].legend()


    axs[i].set_title(pt + "Accuracies on deterministic dataset with hue shift " + str(hue_shift))
    axs[i].set_ylabel("Accuracy")
    axs[i].set_xlabel("Trained on Gaussian dataset with std");
```
