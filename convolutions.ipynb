```python
import torch
from torch import nn
import torch.nn.functional as F
import torchvision
import urllib
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np
```

```python
m = nn.Conv2d(3, 6, kernel_size=3, stride=1, padding=1)
input = torch.randn(3, 5, 5).unsqueeze(0)
output = m(input)
```

```python
m = nn.Conv3d(1, 6, kernel_size=(3,3,3), stride=1, padding = (1, 1, 1))
# non-square kernels and unequal stride and with padding
# m = nn.Conv3d(16, 33, (1, 3, 3), stride=(2, 1, 1), padding=(4, 2, 0))
input = torch.randn(1, 3, 5, 5).unsqueeze(0)
output = m(input)
```

```python
m = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
input = torch.zeros(5, 5)
for i in range(5):
    for j in range(5):
        input[i, j] = 100 - ((i + 1) * (j + 1) * 5)
input = F.pad(input, (2, 2, 2, 2))
input = input.unsqueeze(0)
output = m(input)
```

```python
m = nn.MaxPool3d(kernel_size=(1, 2, 2), stride=(1, 2, 2))
input = torch.zeros(3, 11, 11)
for i in range(3):
    for j in range(11):
        for k in range(11):
            if i == 0:
                input[i, j, k] = 1
            if i == 1:
                input[i, j, k] = 2
            if i == 2:
                input[i, j, k] = 3
                
input = input.unsqueeze(0).unsqueeze(0)
output = m(input)
asd = m(output)
asd1 = m(asd)
```
